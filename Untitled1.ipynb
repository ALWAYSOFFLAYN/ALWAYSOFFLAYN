```python
from datasets import load_dataset
import numpy as np
import torch, gc
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from copy import deepcopy
import random
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

ds = load_dataset("imdb")
train_texts = ds["train"]["text"]
train_labels = np.array(ds["train"]["label"])
test_texts  = ds["test"]["text"]
test_labels = np.array(ds["test"]["label"])
len(train_texts), len(test_texts)
```

installation of **TinyLlama-1.1B** ( model freezed; which means it just gives us a vector embedding for each review from IMBD database )

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = "cuda" if torch.cuda.is_available() else "cpu"
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    load_in_8bit=True,         # требует bitsandbytes - могут вохникнуть сложности из-за установок зависимостей; в Colab же работает из коробки ¯\_(ツ)_/¯
)
model.eval()

@torch.no_grad()
def encode_texts(texts, max_len=256, batch_size=16):
    """
    Получаем эмбеддинг текста как среднее по токенам из последнего скрытого слоя
    базовой части модели (decoder-only).
    """
    embs = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        toks = tokenizer(
            batch, return_tensors="pt", padding=True, truncation=True, max_length=max_len
        )
        toks = {k: v.to(model.device) for k, v in toks.items()}
        # у AutoModelForCausalLM базовая часть доступна как .model
        out = model.model(**toks, output_hidden_states=True)
        last_hid = out.last_hidden_state          # [B, T, H]
        pooled = last_hid.mean(dim=1)             # [B, H] — простое усреднение
        embs.append(pooled.detach().to("cpu"))
        del out, last_hid, toks
        torch.cuda.empty_cache(); gc.collect()
    return torch.cat(embs, dim=0).numpy()
```

---


```python
N_TRAIN = 8000
N_TEST  = 3000

idx_tr = np.random.choice(len(ds["train"]), size=N_TRAIN, replace=False).tolist()
idx_te = np.random.choice(len(ds["test"]),  size=N_TEST,  replace=False).tolist()

sub_tr = ds["train"].select(idx_tr)
sub_te = ds["test"].select(idx_te)

Xtr_text = sub_tr["text"]
ytr = np.array(sub_tr["label"])

Xte_text = sub_te["text"]
yte = np.array(sub_te["label"])
```

```python
Xtr_emb_clean = encode_texts(Xtr_text, max_len=192, batch_size=12)
Xte_emb_clean = encode_texts(Xte_text,  max_len=192, batch_size=12)

print("Формы эмбеддингов:", Xtr_emb_clean.shape, Xte_emb_clean.shape)
```

```python
clf = LogisticRegression(max_iter=2000, n_jobs=-1)
clf.fit(Xtr_emb_clean, ytr)
base_acc = accuracy_score(yte, clf.predict(Xte_emb_clean))
print(f"Базовая точность (чистое обучение, чистый тест): {base_acc:.4f}")
```
